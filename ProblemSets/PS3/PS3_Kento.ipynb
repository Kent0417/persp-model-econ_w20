## Exercise 1.

If the individual lives for one period $T = 1$, she optimizes the following utility function:

$$
\begin{align}
\max_{c_1 \in [0,w_1]} u(c_1) ~~~ \text{s.t. } W_2 = W_1 - c_1
\end{align}
$$

This function can be rewritten:

$$
\begin{align}
\max_{W_2 \in [0,w_1]} u(W_1 - W_2)
\end{align}
$$

Then, the optimal solution under the function is

$$
\begin{align}
W_2 = \min(W_2) = 0 ~~ \forall~W_1
\end{align}
$$

## Exercise 2.

If the individual lives for two periods $T = 2$, she optimizes the following utility function:

$$
\begin{align}
\max_{c_t \in [0,W_t]} \sum_{t = 1}^{T = 2} \beta^{t-1} u(c_t) ~~~ \text{s.t. } W_{t+1} = W_t - c_t
\end{align}
$$

This function can be rewritten:

$$
\begin{align}
\max_{W_2,W_3} u(W_1 - W_2) + \beta u(W_2 - W_3)
\end{align}
$$

The condition of the optimization for the next period $W_3$ in period 2 is

$$
\begin{align}
W_3 = \min(W_3) = 0 ~~ \forall~W_2, W_1
\end{align}
$$

The condition of the optimization for the next period $W_2$ in period 1 is

$$
\begin{align}
                 & \frac{du(c_1)}{dc_1} \cdot \frac{d(W_1 - W_2)}{dW_2} + \beta \frac{du(c_2)}{dc_2} = 0 \\
\Leftrightarrow~ & \frac{du(c_1)}{dc_1} = \beta \frac{du(c_2)}{dc_2}
\end{align}
$$

## Exercise 4.
The optimization problem in period $T-1$ under optimally chosen $W_{T+1}$ is

$$
\begin{align}
V_{T-1}(W_{T-1}) &\equiv \max_{W_T} u(W_{T-1} - W_T) + \beta V_T(W_T) \\
                 & =     \max_{W_T} u(W_{T-1} - W_T) + \beta u(W_T)
\end{align}
$$

The condition that characterizes the optimal choice in period $T - 1$ is

$$
\begin{align}
&\frac{d V_{T-1}(W_{T-1})}{d W_T} = \frac{du(c_T)}{dc_T} \cdot \frac{d(W_{T-1} - W_T)}{dW_T} + \beta \frac{du(W_T)}{dc_T} = 0\\
\Leftrightarrow& \frac{du(W_{T-1} - W_T)}{dc_T} = \beta \frac{du(W_T)}{dc_T} 
\end{align}
$$

Then, the value function $V_{T-1}$ in terms of $\psi_{T-1}(W_{T-1})$ is

$$
V_{T-1}(W_{T-1}) = u(W_{T-1} - \psi_{T-1}(W_{T-1})) + \beta u(\psi_{T-1}(W_{T-1}))
$$

## Exercise 7.
The analytical solution for $\psi_{T-s}(W_{T-s})$ by using induction from the previous exercises is

$$
\psi_{T-s}(W_{T-s}) = \frac{\sum_{i = 1}^{s} \beta^i }{\sum_{k = 0}^{s} \beta^k} W_{T-s}
$$

The analytical solution for $V_{T-s}(W_{T-s})$ by using induction from the previous exercises is

$$
\begin{align}
V_{T-s}(W_{T-s}) = \sum_{k=0}^{s} \beta^{k} \ln\left( \frac{\beta^{k}}{\sum_{i = 0}^{s} \beta^i}  W_{T-s}\right)
\end{align}
$$

Since $\beta \in (0,1)$, the limit of $\psi_{T-s}(W_{T-s})$ as the horizon $s$ approaches infinity is

$$
\begin{align}
\lim_{s \rightarrow \infty} \psi_{T-s}(W_{T-s}) 
    &= \lim_{s \rightarrow \infty} \frac{\sum_{i = 1}^{s} \beta^i }{\sum_{k = 0}^{s} \beta^k} W_{T-s} \\
    &= \left( \frac{\frac{\beta}{1-\beta}}{\frac{1}{1-\beta}} \right) W_{T-s} = \beta W_{T-s} = \psi(W_{T-s})
\end{align}
$$

Also, the limit of $V_{T-s}(W_{T-s})$ as the horizon $s$ approaches infinity is

$$
\begin{align}
\lim_{s \rightarrow \infty} V_{T-s}(W_{T-s}) 
    &= \lim_{s \rightarrow \infty} \sum_{k=0}^{s} \beta^{k} \ln\left( \frac{\beta^{k}}{\sum_{i = 0}^{s} \beta^i} W_{T-s}\right) \\
    &= \lim_{s \rightarrow \infty} 
        \left[
            \sum_{k=0}^{s} \beta^{k} \ln\left(\frac{W_{T-s}}{\sum_{i = 0}^{s} \beta^i} \right)
          + \sum_{k=0}^{s} \beta^{k} \ln(\beta^{k})
         \right]
     = \lim_{s \rightarrow \infty} 
        \left[
            \ln\left(\frac{W_{T-s}}{\sum_{i = 0}^{s} \beta^i} \right) \sum_{k=0}^{s} \beta^{k} 
          + \ln(\beta) \sum_{k=0}^{s} k\cdot\beta^{k} 
         \right]\\
    &= \frac{1}{1-\beta} \ln \left((1-\beta) W_{T-s}\right) + \frac{\beta}{(1 - \beta)^2} \ln(\beta) = V(W_{T-s})
\end{align}

```python
beta = 0.9
Wprime, W = np.meshgrid(Wvec,Wvec)
Cmat = np.array([W.ravel() - Wprime.ravel()]).reshape(N, N)
```

Since $c = W - W' \ge 0$, my `u()` function replaces the  negative entries with very tiny number (close to zero, `1e-10`). Note that the utility function is logarithmic, we need to include zero to replace the entries.

```python
def u(cmat):
    c = cmat.copy()
    c_leq0 = c <= 0
    c[c_leq0] = 1e-10
    return np.log(c)
```

Also, the function below (`gen_Vmat()`) generates $V_{T+1}(W')$ function ($N \times N$ matrix) whose entries corresponding to negative $W-W'$ is replaced with very big negative number

```python
def gen_Vmat(v, c, n):
    c_leq0 = c <= 0
    vmat = np.array([v]*n)
    vmat[c_leq0] = -1e7
    
    return vmat
```

Since the period $T$ is the final period, the value function $T_{T+1}(W')$ would be zero.

```python
Vvec0 = np.zeros(N)
Vmat0 = gen_Vmat(Vvec0, Cmat, N)
```

```python
Vvec1 = (u(Cmat) + beta * Vmat0).max(axis = 1)
```

```python
psi1 = Wvec[np.argmax(u(Cmat) + beta * Vmat0, axis = 1)]
```

```python
from scipy.stats import norm

def gamma(eps, mu, sigma, m):
    prob = []
    mid = [(eps[i] + eps[i+1])/2 for i in range(0, m-1)]
    
    prob.append(norm.cdf(mid[0], mu, sigma))
    ps = [norm.cdf(mid[i], mu, sigma) - norm.cdf(mid[i-1], mu, sigma) for i in range(1,m-1)]
    prob.extend(ps)
    prob.append(1 - norm.cdf(mid[-1], mu, sigma))
                
    return np.array(prob)
```

```python
# params
var = 0.25
M = 7
sigma = np.sqrt(var)
mu = 4 * sigma
eps_max = mu + 3 * sigma
eps_min = mu - 3 * sigma
eps = np.linspace(eps_min, eps_max, M)

probs = gamma(eps,mu,sigma,M)
probs
```

```python
def gen_VTensor3(C, Vmat, probvec, params_args):
    '''
    Generate third-order tensor (M * N * N) for V_T+1(W',eps')
    with entries corresponding to negative W-W' is replaced with
    very big negative number
    
    Input
    ----
    C: N * N matrix
        Consumption matrix (W-W')
    
    Vmat: N * M matrix
        Value function matrix (V(W'))
    
    probvec: M * 1
        probability of epsilon
    
    params_args: [N, M]
    
    Return
    ----
    Vtensor3: ndarray (third-order tensor, M * N * N)
    
    '''
    N, M  = params_args
    E_Vvec = np.dot(probvec, Vmat)
    
    C_leq0 = C <= 0
    Vmat = np.array([E_Vvec] * N)
    Vmat[C_leq0] = -1e7
    
    Vtensor3 = np.array([Vmat] * M)
    
    Vtensor3 = Vtensor3
    return Vtensor3
```
